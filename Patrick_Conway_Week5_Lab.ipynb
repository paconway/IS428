```python
import csv
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

# Step 1: Load the CSV file
def load_csv(file_path):
    data = []
    with open(file_path, 'r', newline='') as file:
        csv_reader = csv.reader(file)
        for row in csv_reader:
            data.append(row)
    return data

# Replace 'example.csv' with the actual path of your CSV file
file_path = 'spam_dataset.csv'
csv_data = load_csv(file_path)

# Extract features and labels
emails = [row[0] for row in csv_data]  # Assuming emails are in the first column
labels = [row[1] for row in csv_data]  # Assuming labels (spam or ham) are in the second column

# Step 2: Preprocess the data
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(emails)

# Step 3a: Train a Naive Bayes classifier
X_train_nb, X_test_nb, y_train_nb, y_test_nb = train_test_split(X, labels, test_size=0.2, random_state=42)
nb_classifier = MultinomialNB()
nb_classifier.fit(X_train_nb, y_train_nb)

# Step 4a: Use the classifier to make predictions for Naive Bayes
print("Predictions for Naive Bayes:")
for i, instance in enumerate(X_test_nb):
    y_pred = nb_classifier.predict(instance.reshape(1, -1))
    print(f"Instance {i+1} - Predicted Label: {y_pred[0]}")

# Step 3b: Train a K-Nearest Neighbors classifier
X_train_knn, X_test_knn, y_train_knn, y_test_knn = train_test_split(X, labels, test_size=0.2, random_state=42)
knn_classifier = KNeighborsClassifier(n_neighbors=5)  # You can adjust 'n_neighbors' as needed
knn_classifier.fit(X_train_knn, y_train_knn)

# Step 4b: Use the classifier to make predictions for KNN
print("\nPredictions for K-Nearest Neighbors:")
for i, instance in enumerate(X_test_knn):
    y_pred = knn_classifier.predict(instance.reshape(1, -1))
    print(f"Instance {i+1} - Predicted Label: {y_pred[0]}")

```

```python
import csv
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

# Step 1: Load the CSV file
def load_csv(file_path):
    data = []
    with open(file_path, 'r', newline='') as file:
        csv_reader = csv.reader(file)
        for row in csv_reader:
            data.append(row)
    return data

# Replace 'example.csv' with the actual path of your CSV file
file_path = 'spam_dataset.csv'
csv_data = load_csv(file_path)

# Extract features and labels
emails = [row[0] for row in csv_data]  # Assuming emails are in the first column
labels = [row[1] for row in csv_data]  # Assuming labels (spam or ham) are in the second column

# Step 2: Preprocess the data
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(emails)

# Step 3: Train a K-Nearest Neighbors classifier
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=42)
knn_classifier = KNeighborsClassifier(n_neighbors=5)  # You can adjust 'n_neighbors' as needed
knn_classifier.fit(X_train, y_train)

# Step 4: Use the classifier to make predictions
y_pred = knn_classifier.predict(X_test)

# Evaluate the classifier
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred, zero_division=0)  # Handle zero division

print(f'Accuracy: {accuracy:.2f}')
print('Classification Report:')
print(report)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

# Load the dataset
data = pd.read_csv('spam_dataset.csv')  # Replace with the actual path of your CSV file
X = data[['word1', 'word2', 'word3', 'word4']]
y = data['spam']

# Define the parameter grid for max_depth
param_grid = {'max_depth': range(1, 11)}

# Initialize a decision tree classifier
dt_classifier = DecisionTreeClassifier(random_state=42)

# Initialize GridSearchCV
grid_search = GridSearchCV(dt_classifier, param_grid, cv=5, scoring='accuracy')

# Fit the grid search to the data
grid_search.fit(X, y)

# Access the results of the grid search
results = grid_search.cv_results_
max_depth_values = results['param_max_depth']
train_scores = results['mean_train_score'] if 'mean_train_score' in results.keys() else results['mean_fit_time']
cv_scores = results['mean_test_score']

# Create a plot
plt.figure(figsize=(10, 6))
plt.plot(max_depth_values, train_scores, label='Train Score', marker='o')
plt.plot(max_depth_values, cv_scores, label='CV Score', marker='o')
plt.xlabel('Max Depth')
plt.ylabel('Accuracy')
plt.title('Max Depth vs. Accuracy')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

```
